```python
# checking blank values for each column
print('title = %d' % (huffpost['headline'].values == '').sum())
print('url = %d' % (huffpost['url'].values == '').sum())
```

```python
# checking tap values for each column
print('title = %d' % (huffpost['headline'].values == ' ').sum())
print('url = %d' % (huffpost['url'].values == ' ').sum())
```

```python
# checking 'na' values for each column
print('title = %d' % (huffpost['headline'].values == 'na').sum())
print('url = %d' % (huffpost['url'].values == 'na').sum())
```

```python
# checking duplicate values for url column
print('url = %d' % (huffpost['url'].duplicated()).sum())
```
